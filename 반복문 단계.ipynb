```python
# 영수증
import sys
x = int(input())
n = int(input())
total = 0
for i in range(n):
    a, b = map(int, sys.stdin.readline().split())
    total += (a*b)

if total == x:
    print("Yes")
else:
    print("No")
```

```python
# 코딩은 체육과목입니다. (20230223 추가)

# 문제
# long int는 4바이트 정수, long 이 하나 늘때 마다 4바이트 추가됨
# N바이트 정수까지 저장할 수 있다고 생각해서 칠판에 쓴 정수 자료형의 이름 ?
# 예) 4 > long int, 20 > long long long long long int

# 풀이
# 정수로 입력 받아 4로 나눈 몫을 이용
# N 은 4이상 1000이하 4의 배수

N = int(input())
c = N//4

result = ["long"]*c
result.append("int")
print(*result, sep = " ")
```

```python
# 더하기 사이클
c = 0
n = int(input())
n_tmp = n
while True:
    a, b = n_tmp//10, n_tmp%10
    n_ = b*10 + (a+b)%10
    c += 1
    if n == n_:
        break
    n_tmp = n_
print(c)
```
