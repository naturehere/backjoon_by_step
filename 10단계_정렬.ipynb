```python
# 수 정렬하기
N = int(input())
nums = [int(input()) for _ in range(N)]
nums.sort()
print(*nums, sep="\n")
```

```python
# 수 정렬하기
N = int(input())
nums = [int(input()) for _ in range(N)]
base = [False]*2001
for i in nums:
    base[i+1001] = True
result = [i - 1001 for i in range(2001) if base[i]]
print(*result, sep = "\n")
```

```python
# 수 정렬하기 3
# 카운팅 정렬을 사용하기

N = int(input())
# nums = [int(input()) for _ in range(N)]
# N = 9
# nums = [4, 7, 9, 1, 3, 5, 2, 3, 4]
num_max = max(nums)
print(nums)

count = [0]*(num_max +1)
# 카운팅
for num in nums:
    count[num] += 1

print(count)
# count 배열 누적 값 합 변환
for i in range(1, len(count)):
    count[i] += count[i-1]
print(count)

# 결과 배열
result = [0]*N

for num in nums:
    idx = count[num]
    result[idx - 1] = num
    count[num] -= 1

print(result)
print(*result, sep='\n')
```

```python
# 수 정렬하기 3
# 카운팅 정렬을 사용하기

N = int(input())
count = [0]*(10001)

# 입력과 카운팅 동시에, 초기 배열을 만들지 않음으로써 메모리 문제 해결.
for num in range(N):
    count[int(input())] += 1

print(count)

# 결과바로 출력
for idx, c in enumerate(count):
    for i in range(c):
        print(idx) 

### 시간초과
```
